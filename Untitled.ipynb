### maximum match vanue

```python
df1=data1.groupby("venue").count().reset_index()
df1=df1[["venue","key"]].sort_values(by=['key'], ascending=False)
df1=df1.head(8)
```

```python
import seaborn as sns
import seaborn as sns
import matplotlib.pyplot as plt
fig, ax = plt.subplots(figsize=(25, 10))
 
# drawing the plot
sns.barplot(df1["venue"],df1["key"], ax=ax)
plt.show()
```

### year of 2012 maximum win team

```python
df2=data1[data1["year"]==2012]
df2=df2.groupby("winner").count().reset_index()
df2=df2[["winner","key"]].sort_values(by=['key'], ascending=False)
df2=df2.head(8)
```

```python
fig, ax = plt.subplots(figsize=(25, 10))
 
# drawing the plot
sns.barplot(df2["winner"],df2["key"], ax=ax)
plt.show()
```

### if you won the toss then how much chance to win the match

```python
xx = data1.shape
xx=xx[0]
count=0
for i in range(xx):
    if data1["toss_winner"][i]==data1["winner"][i]:
        count=count+1
win=(count*100)//xx
print("chance of win the match = ", win,"%" )
```
